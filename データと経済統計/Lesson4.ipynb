# 大数の法則

```
# ダイスを振る回数n
n = 100

# dice = np.array([1, 2, 3, 4, 5, 6])
dice = np.arange(1, 7)
count  = np.arange(1, n + 1)

for i in range(4):
  p = np.random.choice(dice, n).cumsum()
  plt.plot(p / count)
  plt.grid(True)
```

# 中心極限定理

```
set_size = 10000
# dist = np.random.normal(50, 10, set_size)
dist = np.random.poisson(2, set_size)
# dist = np.random.choice(dice, set_size)
# plt.hist(dist)

print('母集団：', dist.mean())
print('母分散：', dist.var(ddof=0))

N = 10000
n = 10
sample_mean = np.array([])

for _ in range(N):
  choice = np.random.choice(dist, n)
  # choice.mean()
  sample_mean = np.append(sample_mean, choice.mean())
  # print(x)

print('標本平均の平均：', sample_mean.mean())
print('標本平均の分散：', sample_mean.var(ddof=1))

plt.hist(dist, alpha=0.5)
plt.hist(sample_mean, alpha=0.5)
```

```
from scipy import stats
import math

n = 400 #サンプルサイズ
mean = 28000
variance = 3060
se = variance/math.sqrt(n) #標準誤差

bottom, up = stats.norm.interval(alpha=0.95, loc=mean, scale=se)
print('{:.2f} < x < {:.2f}'.format(bottom, up))

stats.norm.ppf(0.975)


x = np.linspace(-4, 4, 100)
y_pdf = stats.norm.pdf(x)  # 標準正規分布
plt.plot(x,y_pdf)
plt.xlabel('x')
plt.ylabel('p')
plt.title('Standard Normal Distribution')
```
