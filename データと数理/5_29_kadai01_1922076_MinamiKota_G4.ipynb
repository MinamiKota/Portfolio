## 課題1-1 円の面積から円周率を求めてみよう
- 円に内接する長方形を描いてその面積を求める。
    - 長方形の幅を3種類で計算して、それぞれの図と、それぞれの円周率を求める
    - 半径1の円の面積と円周率の関係は？
- 円に内接する長方形を描くのはちょっとむずかしいので、計算を使って迂回しても良い。
    - 既存のπは使わないこと
    - 完璧な図を描けたら加点する


```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
```

```python
def return_lower2(y1,y2):
    if np.sign(y1)*np.sign(y2) == -1: # 長方形の左端と右端で符号が違うときは0
      yi = 0
    elif np.abs(y2) > np.abs(y1):
      yi = y1
    else:
      yi = y2

    return yi
```

```python
fig = plt.figure(figsize=(3,3), dpi=150) # サイズとdpiの指定
ax = fig.add_subplot(1,1,1) 

ax.set_aspect('equal')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_position('zero')
ax.spines['left'].set_position('zero')

c = patches.Circle(xy=(0, 0), radius=1.0, fc='none', ec='red') # 円を作る
ax.add_patch(c) # 円を追加

ax.axis([-1.5,1.5,-1.5,1.5])

def here_function(x):
    y=np.sqrt(1-x**2)
    return y
start = -1 #定義域の左端
last = 1 # 定義域の右端
x0 = np.arange(start,last, 0.01)
y0 = here_function(x0)
haba = 0.2
x = np.arange(start,last, haba)
y = here_function(x)

area_sum = 0 

for i in range(len(x)):
    if x[i] + haba > last:
        continue
    y1 = y[i]
    y2 = here_function(x[i]+ haba)
    yi = return_lower2(y1, y2)    
    
    r = patches.Rectangle(xy=(x[i], 0), width=haba, height=yi, ec='#000000', fc='#EEEEEE')
    r2 = patches.Rectangle(xy=(x[i],0), width=haba, height=-yi, ec='#000000', fc='#EEEEEE')
    ax.add_patch(r)
    ax.add_patch(r2)
    area_sum += yi*haba*2

plt.show() 
print("area_sum: ", area_sum)
```

## オプション課題1-2 モンテカルロ法を使って円周率を求める
- モンテカルロ法を自分で調べて、円周率を何パターンか求め、その時の図と円周率を示す。
- 参考にしたサイトが有る場合は参照を示すこと。

参照：<br>
Python学習講座「Numpy入門 乱数その3 モンテカルロ法の計算例」<br>
https://www.python.ambitious-engineer.com/archives/1394

```python
#円の内側かどうか判定
def is_inner(x, y):
    return x**2+y**2<=1
 
#10の何乗にするか(10^n)
n=2
 
inner_points_cnt = 0
all_points_cnt = 0
#第一象限
X,Y=np.random.rand(2,10**n)
for x, y in zip(X, Y):
    all_points_cnt += 1
    if is_inner(x, y):
        inner_points_cnt += 1
#第二象限
X2,Y2=np.random.rand(2,10**n)
X2-=1
for x, y in zip(X2, Y2):
    all_points_cnt += 1
    if is_inner(x, y):
        inner_points_cnt += 1
#第四象限
X3,Y3=np.random.rand(2,10**n)
Y3-=1
for x, y in zip(X3, Y3):
    all_points_cnt += 1
    if is_inner(x, y):
        inner_points_cnt += 1
#第三象限
X4,Y4=np.random.rand(2,10**n)
X4-=1
Y4-=1
for x, y in zip(X3, Y3):
    all_points_cnt += 1
    if is_inner(x, y):
        inner_points_cnt += 1

pi = (inner_points_cnt / all_points_cnt)*2*2
# 計算結果を表示
print(str(inner_points_cnt)+"/"+str(all_points_cnt)+"*2*2="+str(pi))
 
# 以下matplotlibによる可視化
fig = plt.figure(figsize=(3,3), dpi=150) # サイズとdpiの指定
ax = fig.add_subplot(1,1,1) 

ax.set_aspect('equal')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_position('zero')
ax.spines['left'].set_position('zero')

c = patches.Circle(xy=(0, 0), radius=1.0, fc='none', ec='red') # 円を作る
ax.add_patch(c) # 円を追加

ax.axis([-1.5,1.5,-1.5,1.5])
# ランダムに配置した点
plt.scatter(X, Y, s=1,c="gray")
plt.scatter(X2,Y2, s=1,c="gray")
plt.scatter(X3,Y3, s=1,c="gray")
plt.scatter(X4,Y4, s=1,c="gray")
 
# グラフの表示
plt.show()
```
