```python
df = sym.diff(f, x)
display(df)
```

```python
d2f = sym.diff(df, x)
display(d2f)
```

```python
temp = sym.integrate(d2f,x)
display(temp)
```

```python
C1 = sym.Symbol('C1')
df_I_d2f = sym.integrate(d2f,x) + C1
display(df_I_d2f)
```

```python
C2 = sym.Symbol('C2')
f_I_df = sym.integrate(df_I_d2f,x) + C2
display(f_I_df)
```

```python
display(f)
display(f_I_df)
```

```python
display(f.subs(x,0))
display(f_I_df.subs(x,0))
```

```python
eq1 = sym.Eq(f.subs(x,0),f_I_df.subs(x,0))
print('eq1:')
display(eq1)
```

```python
display(f.subs(x,1))
display(f_I_df.subs(x,1))
```

```python
eq2 = sym.Eq(f.subs(x,1),f_I_df.subs(x,1))

print('eq2:')
display(eq2)
```

```python
sym.solve([eq1,eq2],[C1,C2])
```
