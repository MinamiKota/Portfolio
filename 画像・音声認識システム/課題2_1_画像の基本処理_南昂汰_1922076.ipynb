```python
## Copy and paste regions
pil_im = Image.open('/content/p001.jpg')
box = (400,250,600,450)
region = pil_im.crop(box)
display(region)
```

```python
## Image contours and histograms

from PIL import Image 
from pylab import *

# read image to array conversion to grayscale
im = array(Image.open('/content/p001.jpg').convert('L'))

# create a new figure
figure()

# don’t use colors
gray()

# show contours with origin upper left corner
contour(im, origin='image')
axis('equal')
axis('off')

# The visualization of the (graylevel) image histogram is done using the hist() function.
figure()
hist(im.flatten(),128)
show()
```

```python
from PIL import Image
## Histogram equalization
from PIL import Image
from numpy import *
from PCV.tools import imtools

im = array(Image.open('/content/p001.jpg').convert('L')) 
im2,cdf = imtools.histeq(im)
```

```python
# don’t use colors
gray()

hist(im.flatten(),128)
show()

# add title and show the plot
title('GrayScale Image')
imshow(im)
```

```python
hist(im2.flatten(),128)
show()

# add title and show the plot
title('Histogram Equalized Image')
imshow(im2)
```

```python
## Morphology - counting objects
## http://en.wikipedia.org/wiki/Mathematical_morphology.

from scipy.ndimage import measurements,morphology

# load image and threshold to make sure it is binary
im = array(Image.open('/content/p001.jpg').convert('L')) 
im = 1*(im<128)
labels, nbr_objects = measurements.label(im) 
print ("Number of objects:", nbr_objects)
```
