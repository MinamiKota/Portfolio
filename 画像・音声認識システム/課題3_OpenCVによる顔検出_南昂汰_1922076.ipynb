```python
# Import the necessary libraries
import numpy as np
import cv2 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
haar_cascade_face = cv2.CascadeClassifier('data/haarcascades/haarcascade_frontalface_alt2.xml')

def convertToRGB(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

def detect_faces(cascade, test_image, scaleFactor = 1.1, minNeighbors = 4):
    # create a copy of the image to prevent any changes to the original one.
    image_copy = test_image.copy()
    
    #convert the test image to gray scale as opencv face detector expects gray images
    gray_image = cv2.cvtColor(image_copy, cv2.COLOR_BGR2GRAY)
    
    # Applying the haar classifier to detect faces
    # lessen minNeighbors to detect more faces in case of less certainty
    faces_rect = cascade.detectMultiScale(gray_image, scaleFactor=scaleFactor, minNeighbors=minNeighbors)

    # Let us print the no. of faces found
    print('Faces found: ', len(faces_rect))

    for (x, y, w, h) in faces_rect:
        cv2.rectangle(image_copy, (x, y), (x+w, y+h), (0, 255, 0), 2)
        
    return image_copy
```

```python
#loading image
test_image2 = cv2.imread('/content/1.png')

# Converting to grayscale
test_image_gray = cv2.cvtColor(test_image2, cv2.COLOR_BGR2GRAY)

# Displaying grayscale image
plt.imshow(test_image_gray, cmap='gray')
```

```python
#call the function to detect faces
faces = detect_faces(haar_cascade_face, test_image2, scaleFactor=1.2, minNeighbors=3)

#convert to RGB and display image
plt.imshow(convertToRGB(faces))
```
