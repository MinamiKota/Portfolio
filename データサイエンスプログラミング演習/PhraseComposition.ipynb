# 音の連結
- ド、レ、ミを1秒ずつ順番に鳴らす

```python
import numpy as np
import IPython.display

A = 1
fs = 44100

t = np.arange(0, 1, 1/fs)
#t = np.arange(0, 10, 1/fs)
y1=A*np.sin(2*np.pi*440*2**((0-9)/12)*t)
y2=A*np.sin(2*np.pi*440*2**((2-9)/12)*t)
y3=A*np.sin(2*np.pi*440*2**((4-9)/12)*t) 

y=np.array([])
y=np.append(y,y1)
y=np.append(y,y2)
y=np.append(y,y3)

IPython.display.Audio(y, rate=fs)
```

```python
from scipy.io.wavfile import write

write('test.wav', fs, y)
```
