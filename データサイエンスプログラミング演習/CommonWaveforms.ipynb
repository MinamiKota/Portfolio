# 4つの基本波形

## サイン波(正弦波, sine wave)
$y=A \sin (2\pi n f t)$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import IPython.display

A = 1
fs = 44100
tone=9
f=440*2**((tone-9)/12)

t = np.arange(0, 1, 1/fs)
y=A*np.sin(2*np.pi*f*t)

plt.plot(t, y)
plt.xlim(0, 1/f*2)
plt.show()

IPython.display.Audio(y, rate=fs)
```

## 三角波(triangle wave)
$y = \frac{8A}{\pi^2} \sum_{k=0}^\infty (-1)^k \frac{\sin((2k+1) 2\pi f t)}{(2k+1)^2}$　
- プログラムでは係数などは省略している

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import IPython.display

A = 1
fs = 44100
tone=9
f=440*2**((tone-9)/12)
N = 100

t = np.arange(0, 1, 1/fs)
y = np.zeros(len(t))
for k in range(N):
    y+=(-1)**k*(A/(2*k+1)**2)*np.sin((2*k+1)*2*np.pi*f*t)

plt.plot(t,y)
plt.xlim(0, 1/f*4)
plt.show()
IPython.display.Audio(y, rate=44100)
```

## のこぎり波(saw wave)
$y = \frac{2A}{\pi} \sum_{k=1}^\infty \frac{\sin(2 \pi kft)}{k}$
- プログラムでは係数などは省略している



```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import IPython.display

A = 1
fs = 44100
tone=9
f=440*2**((tone-9)/12)
N = 5000

t = np.arange(0, 1, 1/fs)
y = np.zeros(len(t))
for n in range(1,N):
    y += 2*A/(np.pi*n)*np.sin(n*2*np.pi*f*t)

plt.plot(t, y)
plt.xlim(0, 1/f*4)
plt.show()

IPython.display.Audio(y, rate=fs)
```

## 矩形波(square wave)
$y = \frac{4A}{\pi} \sum_{k=1}^\infty \frac{\sin((2k-1) \pi f t)}{(2k-1)}$
- プログラムでは係数などは省略している

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import IPython.display

A = 1
fs = 44100
tone=9
f=440*2**((tone-9)/12)

N = 30

t = np.arange(0, 1, 1/fs)
y = np.zeros(len(t))
for n in range(1, N):
    y+= (A/(2*n-1))*np.sin((2*n-1)*2*np.pi*f*t)

plt.plot(t,y)
plt.xlim(0, 1/f*4)
plt.show()
IPython.display.Audio(y, rate=44100)
```
