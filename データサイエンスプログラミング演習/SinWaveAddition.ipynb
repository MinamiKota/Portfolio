# 音の重ね合わせ
- 和音や複数の楽器で同時に複数の音を重ねて全てを聴かせることができる
- 例：「ド」と「ミ」と「ソ」を同時にならす場合

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import numpy
import IPython.display

A = 1
fs = 44100

F=np.array([440*2**((0-9)/12),440*2**((4-9)/12),440*2**((7-9)/12)])
t = np.arange(0, 1, 1/fs)
y=0
for f in F:
    y+=A*np.sin(2*np.pi*f*t) 

plt.plot(t, y)
plt.xlim(0, 1/f*20)
plt.show()

IPython.display.Audio(y, rate=fs)
```

# 近い音を重ね合わせて「うなり」を出す

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import numpy
import IPython.display

A = 1
fs = 44100

t = np.arange(0, 1, 1/fs)

y1=A*np.sin(2*np.pi*440*t) # ラ
y2=A*np.sin(2*np.pi*450*t) # ラとシ♭の間
y=y1+y2

plt.plot(t, y)
plt.xlim(0, 1/440*50)
plt.show()

IPython.display.Audio(y, rate=fs)
```
