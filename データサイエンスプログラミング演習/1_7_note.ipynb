```python
import numpy as np
import IPython.display
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
t=np.arange(0,np.pi*2,0.01)
x=np.array(16*(np.sin(t))**3)
y=np.array(13*np.cos(t)-5*np.cos(2*t)-2*np.cos(3*t)-np.cos(4*t))
plt.plot(x,y)
plt.axis("equal")
plt.show()
```
