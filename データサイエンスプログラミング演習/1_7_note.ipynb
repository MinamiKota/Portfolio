```python
import numpy as np
import IPython.display
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
t=np.arange(0,np.pi*2,0.01)
x=np.array(16*(np.sin(t))**3)
y=np.array(13*np.cos(t)-5*np.cos(2*t)-2*np.cos(3*t)-np.cos(4*t))
plt.plot(x,y)
plt.axis("equal")
plt.show()
```

```python
A=1 #振幅
fs=44100 #サンプリング周波数
n=0 #真ん中のド(C3)がn=0
f=440*(2**((n-9)/12)) #真ん中のド(C3)をn=0としたときの基本周波数
t=np.arange(0,1,1/fs) #音の長さ

y=np.array(A*np.sin(2*np.pi*f*t))
IPython.display.Audio(y,rate=fs)
```
